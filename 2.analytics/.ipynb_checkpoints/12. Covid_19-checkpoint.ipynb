# Анализ данных по COVID-19

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv("covid_19_data.csv")
data_1.head(10)
```

```python
data_1.shape
```

```python
data_2 = data_1.drop(["SNo", "Last Update"], axis=1)
data_2.head(10)
```

```python
# проверка на дубликаты
data_2.duplicated(["ObservationDate", "Province/State", "Country/Region"]).sum()
```

```python
# удаление дубликатов
data_3 = data_2.drop_duplicates(["ObservationDate", "Province/State", "Country/Region"])
```

```python
data_3.duplicated(["ObservationDate", "Province/State", "Country/Region"]).sum()
```

```python
data_3["Country/Region"].unique()
```

```python
# исправляем названия стран
data_3.loc[:,"Country/Region"] = data_3.loc[:,"Country/Region"].str.strip(" (),''") #.unique()
```

```python
data_3.loc[:,"Country/Region"].unique()
```

```python
data_3.columns
```

```python
# изучаем страну "Others"
data_3[data_3["Country/Region"] == "Others"]
```

```python
data_3.columns
```

```python
# pd.to_datetime("03.23.2020")
```

```python
pd.to_datetime("03.23.2020").date()
```

```python
# конвертация временных данных в тип TimeStamp
data_3.loc[:,'ObservationDate'] = pd.to_datetime(data_3['ObservationDate'])
```

```python
# извлечение дат из Timestamp
data_3.loc[:,'ObservationDate'] = data_3['ObservationDate'].apply(lambda el: el.date())
```

```python
data_3.head(10)
```

```python
# Анализ

# группируем данные по датам и выводим суммарные значения (по всему миру)
data_3.groupby(["ObservationDate"]).sum()
```

```python
data_3.columns
```

```python
# статистика по странам и датам

data_country = data_3.groupby(["Country/Region","ObservationDate"])[['Confirmed','Deaths','Recovered']].sum()
data_country.head(10)
```

```python
# данные максимумов по странам
data_max = data_country.groupby(["Country/Region"]).max()
data_max.head(10)
```

```python
# выбираем страну или страны
data_max.loc["Russia", :]
```

```python
# сортировка данных по значению столбцов
# по заболевшим
data_max.sort_values(by='Confirmed', ascending=False).head(10)
```

```python
# сортировка данных по погибшим
data_max.sort_values(by='Deaths', ascending=False).head(10)
```

```python
# Визуализация 
```

```python
data_3.head(10)
```

```python
# Анализ

# группируем данные по датам и выводим суммарные значения (по всему миру)
data_date = data_3.groupby(["ObservationDate"]).sum()
data_date.head(10)
```

```python
plt.figure(figsize=[10, 5])

plt.yscale('log')

plt.plot(data_date.index, data_date['Confirmed'], label='Заболевшие')
plt.plot(data_date.index, data_date['Deaths'], label='Погибшие')
plt.plot(data_date.index, data_date['Recovered'], label='Выздоровевшие')

plt.legend()

plt.show()
```

```python
# оценка летальности
plt.figure(figsize=[10, 5])

letal_data = (data_date['Deaths'] / data_date['Confirmed']) * 100

plt.plot(data_date.index, letal_data)

plt.show()
```

```python
# новые случаи заболевания

# функция для вычисления новых случаев
def new_cases_calculate(df, col_idx):
    new_cases = []
    for idx in range(df.shape[0]):
        if idx == 0:
            new_cases.append(df[col_idx][idx])
        else:
            new_cases.append(df[col_idx][idx] - df[col_idx][idx-1])
    return new_cases
```

```python
new_cases_calculate(data_date, 'Confirmed')
```

```python
plt.figure(figsize=[10, 5])

# plt.yscale('log')

plt.plot(data_date.index, new_cases_calculate(data_date, 'Confirmed'), label='Новые случаи')
# plt.plot(data_date.index, data_date['Confirmed'], label='Заболевшие')

plt.legend()

plt.show()
```

```python
plt.figure(figsize=[10,5])

plt.yscale('log')

plt.plot(ykt_data['ObservationDate'], ykt_data['Confirmed'], label = "Заболевшие")
plt.plot(ykt_data['ObservationDate'], ykt_data['Deaths'], label = "Погибшие")
plt.plot(ykt_data['ObservationDate'], ykt_data['Recovered'], label = "Выздоровевшие")

plt.legend()

plt.show()
```

```python
# оценка летальности
plt.figure(figsize=[10, 5])

letal_data = (ykt_data['Deaths'] / ykt_data['Confirmed']) * 100

plt.plot(ykt_data['ObservationDate'], letal_data)

plt.show()
```

```python
# новые случаи заболевания

# функция для вычисления новых случаев
def new_cases_calculate(df, col_idx):
    new_cases = []
    for idx in range(df.shape[0]):
        if idx == 0:
            new_cases.append(df.iloc[col_idx][idx])
        else:
            new_cases.append(df.iloc[col_idx][idx] - df.iloc[idx-1][idx-1])
    return new_cases
```

```python
new_cases_calculate(data_date, 'Confirmed')
```

```python
# новые случаи

plt.figure(figsize=[10, 5])

plt.plot(ykt_data['ObservationDate'], new_cases_calculate(ykt_data, 'Confirmed'))

plt.show()
```
