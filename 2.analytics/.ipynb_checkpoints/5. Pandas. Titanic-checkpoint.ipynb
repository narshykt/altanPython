# Анализ данных по Титанику

```python
# импортирование модулей
import numpy as np
import pandas as pd
```

```python
# загрузка файла с данными
titanic_data = pd.read_csv('titanic.csv')
```

```python
titanic_data
```

```python
# изучаем часть таблицы (по умолчанию 1 - 5 строк)
titanic_data.head(10)
```

```python
# узнаем количество строк и столбцов
titanic_data.shape
```

```python
# более подробная информация по датафрейму
titanic_data.info()
```

```python
# сводные показатели
titanic_data['age'].mean()
```

```python
# минимальный возраст 
titanic_data['age'].min()
```

```python
# максимальный возраст 
titanic_data['age'].max()
```

```python
# показатели по столбцу 'survived' при условии 'age' == 0.42
titanic_data.loc[titanic_data['age'] == 0.42, 'survived']
```

```python
# показатели по столбцу 'survived' при условии 'age' == 80.0
titanic_data.loc[titanic_data['age'] == titanic_data['age'].max(), 'survived']
```

```python
# показатели по столбцу 'survived' при условии 'age', 'pclass', 'sex' == 0.42
titanic_data.loc[titanic_data['age'] == 0.42, ['survived', 'pclass', 'sex']]
```

```python
# показатели по столбцу 'survived' при условии 'age', 'pclass', 'sex' == 80.0
titanic_data.loc[titanic_data['age'] == titanic_data['age'].max(), ['survived', 'pclass', 'sex']]
```

```python
# сводные показатели по столбцу 'age' при условии 'survived' == 1
titanic_data.loc[titanic_data['survived'] == 1, 'age'].mean()
```

```python
# сводные показатели по столбцу 'age' при условии 'survived' == 0
titanic_data.loc[titanic_data['survived'] == 0, 'age'].mean()
```

```python
# сводные показатели по столбцу 'age' при условии 'survived' == 0
titanic_data.loc[titanic_data['survived'] == 0, 'age'].min()
```

```python
# сводные показатели по столбцу 'age' при условии 'survived' == 0
titanic_data.loc[titanic_data['survived'] == 0, 'age'].max()
```

- GroupBy

```python
# 'сгруппировать по'
# операции разбиение, применение, объединение
titanic_data.groupby('survived').mean()
```

```python
# указание нужных слобцов
titanic_data.groupby('survived')[['pclass', 'age']].mean()
```

```python
titanic_data.head()
```

```python
# группировка данных по категории 'who'
titanic_data.groupby('who')[['pclass', 'age']].mean()
```

```python
# группировка данных по несколььким категориям
titanic_data.groupby(['survived', 'who'])[['pclass', 'age']].mean()
```

```python
# группировка данных по несколььким категориям
titanic_data.groupby(['survived', 'who'])[['pclass', 'age']].mean()
```

```python
# группировка данных по несколььким категориям
titanic_data.groupby(['survived', 'sex'])[['pclass', 'age']].mean()
```

```python
# статистика выживших в зависимости от пола
titanic_data.groupby('sex')['survived'].sum()
```

```python
# статистика выживших в зависимости от класса
titanic_data.groupby('pclass')['survived'].sum()
```

```python
titanic_data.groupby(['class', 'who'])['survived'].sum()
```

```python
# применение нескольких агрегируюющих функций
titanic_data.aggregate(['mean', min, np.max])
```

```python
titanic_data[['pclass', 'age']].aggregate(['mean', min, np.max])
```

```python
titanic_data.groupby(['survived', 'sex'])[['pclass', 'age']].aggregate('min', 'max', 'mean')
```
