# Работа с Excel

-создание из датафрейма

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("titanic.csv")
data.head()
```

```python
# группировка данных по 'survived', 'pclass', 'who'
count_data = data.groupby(['survived', 'pclass', 'who'])['who'].count()
count_data
```

```python
# группировка данных по 'survived', 'pclass', 'who' и узнаем средний возраст
mean_data = data.groupby(['survived', 'pclass', 'who'])['age'].mean()
mean_data
```

```python
df = pd.DataFrame({'count': count_data, 'mean': mean_data})
df
```

```python
# создание excel-file
df.to_excel("titanic.xlsx")
```

```python
# создание excel-file с названием листа
df.to_excel("titanic.xlsx", sheet_name="Titanic")
```

```python
# создание excel-file с несколькими листами

# словарь с датафреймами
sheets = {'Titanic_1': data, "Titanic_2": df}

# определение движка для записи нескольких датафреймов
excelWriter = pd.ExcelWriter('BigTitanic.xlsx', engine="xlsxwriter")

excelWriter
```

```python
# sheets["Titanic_1"].to_excel(excelWriter, sheet_name="T1")
# sheets["Titanic_2"].to_excel(excelWriter, sheet_name="T2")

data.to_excel(excelWriter, sheet_name="one")
df.to_excel(excelWriter, sheet_name="two")

excelWriter.save()
```
