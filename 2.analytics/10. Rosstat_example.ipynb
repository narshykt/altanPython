# Анализ данных из Росстат

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# загрузка данных
raw_data = pd.read_excel("cen-god.xls")
raw_data.head()
```

```python
raw_data_2 = raw_data.dropna()
raw_data_2.head()
```

```python
# строку с датами переделалив индексную строчки для столбцов
raw_data_2.columns = raw_data_2.loc[1, :]
raw_data_2.head()
```

```python
raw_data_2 = raw_data_2.drop(1)
raw_data_2.head()
```

```python
raw_data_2.columns
```

```python
col_arr = np.array(raw_data_2.columns)
col_arr
```

```python
col_arr[col_arr == 2013] = 2013
col_arr = col_arr.astype("str")
col_arr
```

```python
raw_data_2.columns = col_arr
raw_data_2.head()
```

```python
# транспонирование
data_1 = raw_data_2.T
data_1
```

```python
data_1.columns = data_1.loc[" ПОКАЗАТЕЛИ ", :]
data_1
```

```python
data_1 = data_1.drop(" ПОКАЗАТЕЛИ ")
data_1
```

```python
data_1.to_excel("rosstat_data.xlsx")
```

```python
data_1.values
```

```python
data_1.loc["1995", :].unique()
```

```python
data_1.loc["2014", :].unique()
```

```python
# замена символа '…' на 0.0
data_2 = data_1.replace('…', 0.0)
data_2.loc["1995", :].unique()
```

```python
# замена символа '-' на 0.0
data_2 = data_2.replace('-', 0.0)
data_2.loc["2014", :].unique()
```

```python
# замена символа '...' на 0.0
data_2 = data_2.replace('...', 0.0)
```

```python
data_2.values
```

```python
object_cols = list(data_2.select_dtypes(include="object"))
object_cols
```

```python
# замена запятой внутри числа на точку
data_2[object_cols] = data_2[object_cols].replace(',', '.', regex=True)
data_2["Общий анализ крови33)"].unique()
```
