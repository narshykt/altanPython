# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

- простые линейные графики

```python
# 1-мерный массив из 10 значений
x = np.arange(100)
x
```

```python
y = x * 2
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
# 1-мерный массив из 1000 значений
# линейная последовательность в диапазоне от 0 до 10
x = np.linspace(0, 10, 1000)
x
```

```python
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-10, 10, 10)
y = np.tanh(x)

plt.plot(x, y)
plt.show()
```

```python
# def sigmoid(x):
#     return 1 / (1 + np.exp(-x))

sigmoid = lambda a: 1 / (1 + np.exp(-a))

y = sigmoid(x)

plt.plot(x, y)
plt.show()
```

```python
# совмещение двух графиков в одном окне
x = np.linspace(-10, 10, 1000)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.cos(x - 1)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
# настройка графиков: цвет и вид линий

# цвет можно задавать разными способами
plt.plot(x, np.cos(x), color="red") # по названию
plt.plot(x, np.cos(x + 1), color="m") # по первой букве цвета из rgbcmyk
plt.plot(x, np.cos(x + 2), color="#ffdd4f") # по 16-ричному коду RGB
plt.plot(x, np.cos(x + 3), color=(0.5, 1.0, 0.8)) # RGB-кортеж со значениями в диапазоне от 0 до 1

plt.show()
```

```python
# стиль линий
plt.plot(x, np.cos(x), linestyle="solid") #'-') # сплошная
plt.plot(x, np.cos(x - 1), linestyle="dashed") #'--') # штриховая
plt.plot(x, np.cos(x - 2), linestyle=':') # пунктирная
plt.plot(x, np.cos(x - 3), linestyle='-.') # штрих-пунктирная

plt.show()
```

```python
# объединение цветов и стилей линий
plt.plot(x, np.cos(x), linestyle='-', color="r") # сплошная красная
plt.plot(x, np.cos(x - 1), linestyle='--', color="g") # штриховая зеленая
plt.plot(x, np.cos(x - 2), linestyle=':', color="b") # пунктирная голубая

plt.show()
```

```python
# объединение цветов и стилей линий
plt.plot(x, np.cos(x), "-g") # сплошная зеленая
plt.plot(x, np.cos(x - 1), '--k') # штриховая черная
plt.plot(x, np.cos(x - 2), ':b') # пунктирная голубая

plt.show()
```

```python
# пределы осей

plt.plot(x, np.cos(x), "-g")

# индивидуальные настройки осей
# plt.xlim(-15, 15)
# plt.ylim(-2, 2)

plt.axis([-20, 20, -1.5, 1.5])

plt.show()
```

```python
# легенда графика
plt.plot(x, np.cos(x), "-r", label="косинусоида")
plt.plot(x, np.sin(x), "-g", label="синусоида")

plt.legend()

plt.axis([-20, 20, -1.5, 1.5])

plt.show()
```

```python
# подписи осей
plt.plot(x, np.cos(x), "-r", label="cos")
plt.plot(x, np.sin(x), "-g", label="sin")

plt.legend()
plt.xlabel("x")
plt.ylabel("sin and cos")

plt.axis([-20, 20, -1.5, 1.5])

plt.show()
```

```python
# заголовок графиков
plt.plot(x, np.cos(x), "-r", label="cos")
plt.plot(x, np.sin(x), "-g", label="sin")

plt.legend()
plt.xlabel("x")
plt.ylabel("sin and cos")
plt.title("Графики синуса и косинуса")

plt.axis([-20, 20, -1.5, 1.5])

plt.show()
```
